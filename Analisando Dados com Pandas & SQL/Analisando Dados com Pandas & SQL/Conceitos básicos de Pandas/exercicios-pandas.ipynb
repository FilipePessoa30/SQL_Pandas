### 4. Você deve ter percebido que as colunas DATA INICIAL e DATA FINAL estão formatadas como string. Utilizando o método pd.to_datetime(), converta ambas colunas para Timestamp / Datetime.

```python
df1['DATA INICIAL'] = pd.to_datetime(df1['DATA INICIAL'])
df1['DATA FINAL'] = pd.to_datetime(df1['DATA FINAL'])
df2['DATA INICIAL'] = pd.to_datetime(df2['DATA INICIAL'])
df2['DATA FINAL'] = pd.to_datetime(df2['DATA FINAL'])
```

### 5. Crie uma nova coluna para representar o ano e o mês(aaaa-mm), utilizando a coluna DATA FINAL como referência.

### 7. Filtre o DataFrame para obter apenas dados da 'GASOLINA COMUM'. Grave em uma nova variável.

```python
gasolina_comum = pd.concat([df1[df1['PRODUTO'] == 'GASOLINA COMUM'], df2[df2['PRODUTO'] == 'GASOLINA COMUM']])
```

```python
gasolina_comum
```

### 8. Qual o preço médio de revenda da gasolina em agosto de 2008?

```python
filtro = gasolina_comum[gasolina_comum['ANO-MES'] == '2008-08']

filtro['PREÇO MÉDIO REVENDA'].mean()
```

### 9. Qual o preço médio de revenda da gasolina em maio de 2014 em São Paulo?

```python
filtro = gasolina_comum_[gasolina_comum_['ANO-MES'] == '2014-05']

filtro['PREÇO MÉDIO REVENDA'].mean()
```

### 10. Você conseguiria descobrir em qual(quais) estado(s) a gasolina ultrapassou a barreira dos R$ 5,00? E quando isso ocorreu?

```python
filtro = gasolina_comum[gasolina_comum['PREÇO MÉDIO REVENDA'] > 5.00]

filtro['ESTADO'].unique()
```

### 11. Qual a média de preço dos estados da região sul em 2012?

```python
gasolina_comum.info()
```

```python

filtro = gasolina_comum[(gasolina_comum['REGIÃO'] == 'SUL') & (gasolina_comum['DATA FINAL'].dt.year == 2012)]

filtro['PREÇO MÉDIO REVENDA'].mean()
```

### 12. Você conseguiria obter uma tabela contendo a variação percentual ano a ano para o estado do Rio de Janeiro?

```python
RJ = gasolina_comum[gasolina_comum['ESTADO'] == 'RIO DE JANEIRO']
```

```python
filtro = RJ.groupby(RJ['DATA FINAL'].dt.year)['PREÇO MÉDIO REVENDA'].mean().sort_index()
```

```python
filtro.pct_change() * 100
```
