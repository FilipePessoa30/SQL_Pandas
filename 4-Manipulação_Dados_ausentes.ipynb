# Manipulação de dados ausentes

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({
    'A': [1, 2, np.nan],
    'B': [5, np.nan,np.nan],
    'C': [1,2,3]})
    
```

```python
df
```

```python
df.dropna()
```

```python
df.dropna(axis=1)
```

```python
df.dropna(axis=0)
```

```python
df.dropna(axis=1, thresh=2)
```

```python
df
```

```python
df.fillna(0)
```

```python
df.fillna('Conteúdo')
```

```python
df['A'].mean()
```

```python
df['A'  ].fillna(df['A'].mean())
```

```python
df.ffill()
```

```python
df
```

```python
df.bfill() # Preencher com último valor conhecido
```

# Groupby

```python
import pandas as pd
#Cria um DataFrame
data = {'Classe':['Júnior', 'Júnior', 'Pleno', 'Pleno', 'Sênior', 'Sênior'],
        'Nome': ['Jorge', 'Carlos', 'Roberta', 'Patrícia', 'Bruno', 'Vera'],
        'Venda' : [200,120, 340, 124, 243, 350]}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
group = df.groupby('Classe')
```

```python
df.groupby("Classe").sum(numeric_only=True)
```

```python
df.groupby("Classe").mean(numeric_only=True)
```

```python
df.groupby("Classe").min()
```

```python
df.groupby("Classe").max()
```

```python
df2 = df.copy()
```

```python
df2
```

```python
df2['Venda'] = [150, 432, 190, 230, 410, 155]
```

```python
df
```

```python
df2
```

```python
df3 = pd.concat([df, df2], ignore_index=True)
```

```python
df3
```

```python
df3.groupby(["Classe", "Nome"]).sum()
```
